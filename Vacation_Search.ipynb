```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
city_df = pd.read_csv("WeatherPy_challenge.csv")
city_df.head()
```

```python
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))

# Ask the customer to indicate rain preference and snow preference
raining = input("Would you like to be raining? (yes/no)")
snowing = input("Would you like it to be snowing? (yes/no)")
```

```python
# Setting up Filtering
for index, row in city_df.iterrows():
    if raining == 'yes':
        rainfall = city_df["Rainfall (in) past three hours"] != 0
    else:
        rainfall = city_df["Rainfall (in) past three hours"] == 0
    if snowing == "yes":
        snowfall = city_df["Snowfall (in) past three hours"] != 0
    else:
        snowfall = city_df["Snowfall (in) past three hours"] == 0
```

```python
# Filter the dataset to find the cities that fit the criteria.
preferred_cities_df = city_df.loc[(city_df["Max Temp"] <= max_temp) & (city_df["Max Temp"] >= min_temp) & 
                                  (city_df["Rainfall (in) past three hours"] != rainfall) & 
                                  (city_df["Snowfall (in) past three hours"] != snowfall)]
#???
preferred_cities_df.head(10)
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = preferred_cities_df[["City", "Country", "Max Temp", "Lat", "Lng", "Weather"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Set the parameters to search for a hotels near locations
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key}
# Use base URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
# Make request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```
